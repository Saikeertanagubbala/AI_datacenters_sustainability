```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("cleaned_data_centers_current.csv")
display(df.head())
```

```python
print(df.info())
df.describe()
print(df.isnull().sum()) #gives a sum of all the misisng values per variable/column in df
print(df.columns)
print(df.shape)
```

```python
columns_to_clean = [
    'hyperscale_data_centers',
    'colocation_data_centers',
    'power_capacity_MW_total',
    'internet_penetration_percent',
    'floor_space_sqft_total',
    'growth_rate_of_data_centers_percent_per_year',
    'average_renewable_energy_usage_percent']

for col in columns_to_clean:
    df[col] = pd.to_numeric(
        df[col].astype(str).str.replace(r'[~+,%]', '', regex=True),
        errors='coerce'
    )
```

```python
print(df.info())
df.describe()
print(df.isnull().sum()) #gives a sum of all the misisng values per variable/column in df
print(df.columns)
```

```python
df.to_csv('cleaned_data_centers_current.csv', index=False)
```

```python
numerical_cols_with_missing = [
    'floor_space_sqft_total', 
    'power_capacity_MW_total', 
    'average_renewable_energy_usage_percent', 
    'internet_penetration_percent',
    'growth_rate_of_data_centers_percent_per_year'
]

complete_numerical_cols = [
    'total_data_centers', 
    'hyperscale_data_centers', 
    'colocation_data_centers'
]

all_numerical_cols = complete_numerical_cols + numerical_cols_with_missing
data_complete = df[all_numerical_cols].dropna()
corr_before = data_complete.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(corr_before, annot=True, fmt='.2f', cmap='coolwarm', 
            center=0, square=True, linewidths=1)
plt.title('Correlation Matrix - Complete Cases Only (Before Imputation)')
plt.tight_layout()
plt.show()

print(f"Complete cases used: {len(data_complete)}/{len(df)}")
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer, SimpleImputer
from sklearn.preprocessing import StandardScaler, RobustScaler
import matplotlib.pyplot as plt

# Columns having missing values 
X = pd.DataFrame(df, columns=['floor_space_sqft_total', 'power_capacity_MW_total', 'average_renewable_energy_usage_percent', 'internet_penetration_percent', 'growth_rate_of_data_centers_percent_per_year'])
X.head()
```

```python
lr = LinearRegression()
imp = IterativeImputer(estimator=lr, verbose=2, max_iter=5, tol=1e-10, imputation_order='roman')
ImputedData = imp.fit_transform(X)
```

```python
Imputed_data = pd.DataFrame(ImputedData)
Imputed_data.columns = ['floor_space_sqft_total', 'power_capacity_MW_total', 'average_renewable_energy_usage_percent', 'internet_penetration_percent', 'growth_rate_of_data_centers_percent_per_year']
Imputed_data.head()
```

```python
completed_data = df.to_csv('cleaned_data_centers_current.csv', index=False)
completed_data = pd.DataFrame(completed_data, columns=['floor_space_sqft_total', 'power_capacity_MW_total', 'average_renewable_energy_usage_percent', 'internet_penetration_percent', 'growth_rate_of_data_centers_percent_per_year'])
```

```python
mse = mean_squared_error(Imputed_data, completed_data)
print(f'Mean Squared Error (MSE): {mse}')

mae = mean_absolute_error(Imputed_data, completed_data)
print(f'Mean Absolute Error (MAE): {mae}')

r_squared = r2_score(Imputed_data, completed_data)
print(f'Coefficient of Determination (R-squared): {r_squared}')
```
